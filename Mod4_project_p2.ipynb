```python
# 6.2.7 example: word counts
# Example 6.2.7 shows a common task in big data - what are the most common words appearing in a set of text. 
# Create and execute the example from 6.2.7 to perform word count using Python dictionaries. 
# Modify the text to some other source that has more words than the original example. 
# Execute your modified program and show the results. Take a screenshot after executing your modified program. 
# Sets are easy after using the other collections. 
text = ('this is sample text with several words '
        'this is more sample text with some different words')
word_counts = {}

for word in text.split():
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1

print(f'{"WORD":<12}COUNT')

for word, count in sorted(word_counts.items()):
    print(f'{word:<12}{count}')

print('\nNumber of unique words:', len(word_counts))
print()
####################
from collections import Counter

text = ('this is sample text with several words '
        'this is more sample text with some different words')

counter = Counter(text.split())

for word, count in sorted(counter.items()):
    print(f'{word:<12}{count}')

print('\nNumber of unique keys:', len(counter.keys()))
print()
####################
import random

numbers = [random.randrange(1, 6) for i in range(50)]

from collections import Counter

counter = Counter(numbers)

for value, count in sorted(counter.items()):
    print(f'{value:<4}{count}')
print()
```

```python
# 6.3 sets
# Sets are easy after using the other collections. 
# Work through Section 6.3 executing the examples. Take a screenshot when you finish. 
print('6.3 sets')
colors = {'red', 'orange', 'yellow', 'green', 'red', 'blue'}
print(colors)
print(len(colors))
print('red' in colors)
print('purple' in colors)
print('purple' not in colors)
for color in colors:
    print(color.upper(), end=' ')
print()
numbers = list(range(10)) + list(range(5))
print(set(numbers))
    
text = 'to be or not to be that is the question'
unique_words = set(text.split())
for word in sorted(unique_words):
    print(word, end=' ')
print()    
print() 
##########
print('6.3.1 comparing sets')
print({1, 3, 5} == {3, 5, 1})
print({1, 3, 5} != {3, 5, 1})
print({1, 3, 5} < {3, 5, 1})
print({1, 3, 5} < {7, 3, 5, 1})
print({1, 3, 5} <= {3, 5, 1})
print({1, 3} <= {3, 5, 1})
print({1, 3, 5}.issubset({3, 5, 1}))
print({1, 2}.issubset({3, 5, 1}))
print({1, 3, 5} > {3, 5, 1})
print({1, 3, 5, 7} > {3, 5, 1})
print({1, 3, 5} >= {3, 5, 1})
print({1, 3, 5} >= {3, 1})
print({1, 3} >= {3, 1, 7})
print({1, 3, 5}.issuperset({3, 5, 1}))
print({1, 3, 5}.issuperset({3, 2}))
set('abc def ghi jkl mno').issuperset('hi mom')
print()
##########
print('6.3.2 mathematical set operations')
print({1, 3, 5} | {2, 3, 4})
print({1, 3, 5}.union([20, 20, 3, 40, 40]))
print({1, 3, 5} & {2, 3, 4})
print({1, 3, 5}.intersection({1, 2, 2, 3, 3, 4, 4}))
print({1, 3, 5} - {2, 3, 4})
print({1, 3, 5, 7}.difference({2, 2, 3, 3, 4, 4}))
print({1, 3, 5} ^ {2, 3, 4})
print({1, 3, 5, 7}.symmetric_difference({2, 2, 3, 3, 4, 4}))
print({1, 3, 5}.isdisjoint({2, 4, 6}))
print({1, 3, 5}.isdisjoint({4, 6, 1}))
print({10, 20, 30} - {5, 10, 15, 20})
print({10, 20, 30} ^ {5, 10, 15, 20})
print({10, 20, 30} | {5, 10, 15, 20})
print({10, 20, 30} & {5, 10, 15, 20})
print()
##########
print('6.3.3 mutable set operators and methods')
numbers = {1, 3, 5}
numbers |= {2, 3, 4}
print(numbers)
numbers.update(range(10))
print(numbers)
numbers.add(17)
numbers.add(3)
print(numbers)
numbers.remove(3)
print(numbers)
numbers.pop()
print(numbers)
numbers.clear()
print(numbers)
print()
##########
print('6.3.4 set comprehensions')
numbers = [1, 2, 3, 4, 5, 6, 6, 7, 8, 9, 10, 10]
evens = {item for item in numbers if item % 2 == 0}
print(evens)
print()
print('amanda hanway')
```
