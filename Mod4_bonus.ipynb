```python
from collections import Counter

catFile = open('C:/Users/mandi/OneDrive/Documents/NWMSU/44-608/608-mod4/cat.txt', encoding='utf8')
catText = catFile.read()
catFile.close()
# print(catText)

counter = Counter(catText.split())
# print('Sample Text:\n', catText)
# print()

print('\nNumber of unique keys in the cat file:', len(counter.keys()))

for word, count in sorted(counter.items()):
    print(f'{word:<20}{count}')


```
